## Loading breast cancer dataset from sklearn library

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

breast_cancer_data = load_breast_cancer()
```

## Creating a dataframe for the dataset

```python
data = pd.DataFrame(breast_cancer_data.data, columns=breast_cancer_data.feature_names)


data['target'] = breast_cancer_data.target
target = breast_cancer_data.target_names 
data['target'] = data['target'].map(lambda x: target[x])
data
```

## Data Preprocessing

```python
data.info()
data.describe()
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

features = data.drop('target', axis=1)
target = data['target']

x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=.2, random_state=42)
                                                    
scaler = StandardScaler()                                                  
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

### Preprocessing Steps

#### 1. Loading the Dataset:
- The Breast Cancer dataset was loaded using the `load_breast_cancer()` function from scikit-learn.

#### 2. Converting to a DataFrame:
- The dataset was converted into a pandas DataFrame for easier manipulation.

#### 3. Mapping Target Values:
- Numeric target values (`0` for malignant, `1` for benign) were mapped to their corresponding string labels ('malignant' and 'benign').

#### 4. Separating Features and Target:
- The features (independent variables) were separated from the target (dependent variable) for further processing.

#### 5. Feature Scaling Using StandardScaler:
- Features were standardized using `StandardScaler` to ensure that all features contribute equall to the model.


## Model building

### Classifiacation Algorithms
#### 1. Logistic Regreassion

```python
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression()

logistic_model.fit(x_train_scaled, y_train)
y_pred_logistic = logistic_model.predict(x_test_scaled)
```

#### 2. Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier(random_state=42)

tree_model.fit(x_train_scaled, y_train)
y_pred_tree = logistic_model.predict(x_test_scaled)
```

#### 3. Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(random_state=42, n_estimators=100)

random_forest.fit(x_train_scaled, y_train)
y_pred_rf = logistic_model.predict(x_test_scaled)
```

#### 4. Support Vector Machine (SVM)

```python
from sklearn.svm import SVC

svm_model = SVC(kernel='linear', random_state=42) 

y_pred_svc = svm_model.fit(x_train_scaled, y_train)
y_pred_svc = logistic_model.predict(x_test_scaled)
```

#### 5. k-Nearest Neighbors (k-NN)

```python
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
import matplotlib.pyplot as plt

knn_model = KNeighborsClassifier(n_neighbors=5)

knn_model.fit(x_train_scaled, y_train)
y_pred_knn = logistic_model.predict(x_test_scaled)
```

## Comparing performance of the models

### Accuracy score, Classification report, Confusion matrix
#### 1. Logistic Regreassion

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

target_names = breast_cancer_data.target_names
```

```python
accuracy_logistic = accuracy_score(y_test, y_pred_logistic)
print(classification_report(y_test, y_pred_logistic, target_names=target_names))
cofusion_matrix_logistic = confusion_matrix(y_test, y_pred_logistic, labels=target_names)
sns.heatmap(cofusion_matrix_logistic, annot=True, fmt='d', cmap='Blues', xticklabels=target_names, yticklabels=target_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix Heatmap of Logistic Regression')
plt.show()
```

#### 2. Decision Tree Classifier

```python
accuracy_tree = accuracy_score(y_test, y_pred_tree)
print(classification_report(y_test, y_pred_tree, target_names=target_names))
cofusion_matrix_tree = confusion_matrix(y_test, y_pred_tree, labels=target_names)
sns.heatmap(cofusion_matrix_tree, annot=True, fmt='d', cmap='Blues', xticklabels=target_names, yticklabels=target_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix Heatmap of Decision Tree Classifier')
plt.show()
```

#### 3. Random Forest Classifier

```python
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(classification_report(y_test, y_pred_rf, target_names=target_names))
cofusion_matrix_rf = confusion_matrix(y_test, y_pred_rf, labels=target_names)
sns.heatmap(cofusion_matrix_rf, annot=True, fmt='d', cmap='Blues', xticklabels=target_names, yticklabels=target_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix Heatmap of Random Forest Classifier')
plt.show()
```

#### 4. Support vector Machine (SVM)

```python
accuracy_svc = accuracy_score(y_test, y_pred_svc)
print(classification_report(y_test, y_pred_svc, target_names=target_names))
cofusion_matrix_svc = confusion_matrix(y_test, y_pred_svc, labels=target_names)
sns.heatmap(cofusion_matrix_svc, annot=True, fmt='d', cmap='Blues', xticklabels=target_names, yticklabels=target_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix Heatmap of Support Vector Machine(SVM)')
plt.show()
```

#### 5. K-Nearest Neighbors (KNN)

```python
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print(classification_report(y_test, y_pred_knn, target_names=target_names))
cofusion_matrix_knn = confusion_matrix(y_test, y_pred_knn, labels=target_names)
sns.heatmap(cofusion_matrix_knn, annot=True, fmt='d', cmap='Blues', xticklabels=target_names, yticklabels=target_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix Heatmap of k-Nearest Neighbors (KNN)')
plt.show()
```

### Finding best algorithm and worst algorithm

```python
accuracies = {
    'Logistic Regression': accuracy_logistic,
    'Decision Tree': accuracy_tree,
    'Random Forest': accuracy_rf,
    'SVM': accuracy_svc,
    'k-NN': accuracy_knn
}

best_model = max(accuracies, key=accuracies.get)
worst_model = min(accuracies, key=accuracies.get)

print(f"\nBest Performing Model: {best_model} with Accuracy: {accuracies[best_model]:.4f}")
print(f"Worst Performing Model: {worst_model} with Accuracy: {accuracies[worst_model]:.4f}")
```
